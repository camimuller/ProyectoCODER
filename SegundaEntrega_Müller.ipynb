A continuación se determinan los valores atipicos de dicha variable a partir del Método de las desviaciones absolutas medianas (MAD), el cual es ideal para aplicar en distribuciones asimétricas.

```python
# Calcular las desviaciones absolutas medianas
median = np.median(df.glucose)
mad = np.median(np.abs(df.glucose - median))

# Definir umbral para los outliers por lo general se usa desde 2,5 hasta 3,5
threshold = 3.5

# Detección de outliers
outliers = df.totChol[np.abs(df.glucose- median) / mad > threshold]

#veamos los resultados finales
print("Outliers detectados por el Método de las Desviaciones Absolutas Medianas (MAD):")
print(outliers)
```

En este caso en particular, los outliers son representativos de un subconjunto particular de los datos y contienen información valiosa, por lo que se opta por mantenerlos en el conjunto de datos y tenerlos en cuenta en el análisis.

El rango intercuartílico (IQR=Q3-Q1) es una medida de variabilidad, basada en la división de un conjunto de datos en cuatro partes iguales. El 50% de los datos está dentro de este rango IQR y a mayor IQR significa que los datos están mas dispersos. Los datos que quedan fuera de Q3+1,5xIQR o de Q1-1,5xIQR se consideran valores atípicos. De las variables analizadas anteriormente se puede concluir que los datos de Colesterol total son los mas dispersos ya que presentan mayor IQR y los de Indice de masa corporal son los menos dispersos de la base de datos debido a que presentan el menor IQR.

### ***Clustering (K-means)***

Se implementa el algoritmo K-means con el propósito de determinar si existen relaciones entre las variables de la base de datos, para ello es importante seleccionar las variables numéricas adecuadas para el análisis. K-means es un algoritmo de clustering que agrupa los datos en "K" grupos basados en la similitud de las variables, dado que es un algoritmo no supervisado, no requiere una variable de predicción (target), sino que se enfoca en encontrar patrones y agrupaciones naturales en los datos.

En el contexto de la base de datos analizada, donde el objetivo es predecir el riesgo de enfermedades coronarias, se pueden considerar las variables médicas que están relacionadas con la salud de los pacientes para implementar el algoritmo K-means: Edad, nivel de colesterol total, Presión arterial sistólica y diastólica, Índice de masa corporal y Nivel de glucosa. Estas variables son factores importantes, directamente relacionadas con el riesgo a padecer enfermedades coronarias y podrían contribuir a la formación de grupos de pacientes con riesgos similares.

Para aplicar el algoritmo K-means a variables categóricas como el sexo (masculino/femenino) o si es fumador o no (sí/no), se deben convertir estas variables en variables binarias del tipo 0 y 1, lo que se puede realizar aplicando One Hot Encoding. Sin embargo, se debe tener en cuenta que K-means es más adecuado para variables numéricas y continuas, y su aplicación directa a variables categóricas binarias puede tener limitaciones debido a que el algoritmo lo que hace es minimizar la varianza entre datos del mismo cluster y maximizar la varianza entre datos de diferentes cluster.




```python
# se crea el dataframe con los atributos a analizar de la base de datos original
df_kmeans = df.drop(columns=["sex", "education", "is_smoking","cigsPerDay", "BPMeds", "prevalentHyp","prevalentStroke", "diabetes", "heartRate", "TenYearCHD"])
df_kmeans.head()
```

```python
model = KMeans()   #determinación del hiperparámetro k a partir de la regla del codo. La librería yellowbrick tiene un método rápido para visualizarlo.
visualizer = KElbowVisualizer(model, k=(1,12))
visualizer.fit(df_kmeans)        # Entrenamos con los datos
visualizer.show()        # Renderizamos la imagen
```

El número óptimo de clústeres (k) en el algoritmo K-means es 4 debido a que es donde se observa el punto de inflección de la curva azul.

####Escalado de los datos



Se realiza un preprocesamiento para normalizar o escalar las variables, debido a que los resultados obtenidos con K-means no establecen una relación causal entre las variables, sino que agrupan los datos en función de la similitud entre las características seleccionadas por lo que el algoritmo es sensible a la escala de las características.

Mediante la libreria sklearn se construye un pipeline, donde los datos deben ser estandarizados ya que al tener las variables de entrada diferente naturaleza, puede generar ruido en la segmentación. Se utiliza el objeto StandardScaler para estandarizar los datos, es decir, para transformar las características de forma que tengan una media de cero y una desviación estándar de uno.

```python
modelo= Pipeline([('scaler', StandardScaler()),

                  ('model',KMeans(n_clusters=4, max_iter=1000, init='random', random_state=42))
])

modelo.fit_predict(df_kmeans)
preds = modelo.fit_predict(df_kmeans)
```

###**Análisis por componentes principales (PCA)** Reducción de la dimensionalidad

Se busca reducir la cantidad de variables del dataset pero reteniendo la mayor cantidad de información posible generando nuevas variables llamadas componentes. Esto facilita la comprensión de los datos, detectección de las variables mas relevantes o cuales estan altamente correlacionadas, mejora la vizualización los datos, reduce el costo computacional al entrenar algoritmos con menor número de features y mejora el funcionamiento de los modelos evitando la "maldición de la dimensionalidad".

```python
# se crea un nuevo dataframe a partir de las variables númericas del df original.
df_PCA = df[['age', 'cigsPerDay', 'totChol', 'sysBP', 'diaBP', 'BMI', 'heartRate', 'glucose']]
df_PCA.head()
```

```python
modelo_pca = PCA(n_components=8) # Se crea la instancia del modelo

modelo_pca.fit(df_PCA)  # Se entrena el modelo con los datos
```

```python
modelo_pca.components_
```

```python
modelo_pca.explained_variance_ratio_[0]
```

```python
# Se extrae la información del modelo para poderlo interpretar a partir de .components_
# se almacena la nueva información en un dataframe.
lista_componentes = ['PC1', 'PC2', 'PC3', 'PC4','PC5', 'PC6', 'PC7', 'PC8']
componentes = pd.DataFrame(data=modelo_pca.components_, columns = df_PCA.columns, index=lista_componentes)
componentes
```

```python
ax = sns.heatmap(componentes, annot=True) #se grafica para interpretarlo
```

Del gráfico anterior se observa que la nueva componente **PC1** esta fuertemente relacionada con el nivel de colesterol total. Por su parte, **PC2** y **PC3** estan relacionadas fuertemente con el nivel de glucosa y las presiones sistólia y diastólica.
Las componentes **PC4** y **PC5** se relacionan con la cantidad de cigarrillos fumados por día y la frecuencia cardíaca, mientras que **PC6** esta relacionada con la edad del paciente y la presión diastólica. La componente **PC7** se relaciona con la edad del paciente y las presiones sistólica y diastólica. Por último, la **PC8** esta fuertemente relacionada con el índice de masa corporal.


```python
# Se observa la varianza explicada de cada componente para eso se utiliza .explained_variance_ratio_
print(lista_componentes)
print(modelo_pca.explained_variance_ratio_.round(2))
```

Las componente PC1, PC2 y PC3 explican el 90% de la varianza.

```python
# Se observa graficamente la distribución de la varianza en los nuevos componentes.
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(lista_componentes, modelo_pca.explained_variance_ratio_, alpha=0.7, label='Varianza explicada por componente')


ax.set_title('Componentes principales de la base de datos')
ax.set_xlabel('Componente')
ax.set_ylabel('Varianza')
ax.set_ylim(0, 1.2)
ax.legend(bbox_to_anchor = (1.5, 0.6))
```

```python
# se utiliza .transform para calcular la función para cada dato y generar. Dicha información se almacena en un dataframe.
proyecciones = pd.DataFrame(modelo_pca.transform(df_PCA), index=df_PCA.index,columns=lista_componentes)
proyecciones
```

##**Algoritmo de clasificación - Arbol de decisión**

Se seleccionan las variables predictoras que se consideran relevantes para predecir la variable objetivo (en este caso, el riesgo de padecer enfermedades coronarias).
A partir del análisis exploratorio de datos, el cual tambien contiene un algoritmo de clustering (K-means) y  reducción de dimensionaidad (PCA), se realiza la selección de las variables mas relevantes y en el caso de aquellas variables predictorias que estén altamente correlacionadas, se considera solo una de ellas.

```python
# Se define la matriz de diseño X y vector respuesta y. X=features y=target
X = df[['age', 'totChol', 'sysBP', 'BMI', 'glucose']]
X
```

```python
y = df['TenYearCHD'] # target
y
```

```python
y.value_counts()
```

```python
y.value_counts(normalize=True).round(2)
```

Es un problema de clases desbalanceado debido a que el 85% de los registros no tienen riesgo a padecer enfermedades coronarias y el 15% si lo tienen.

```python
# Se separa en entrenamiento/test  (70/30%)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
clf = DecisionTreeClassifier(max_depth=6) # Se crear la instancia del modelo

# Se entrena el modelo con los datos y la etiqueta (target)
clf = clf.fit(X_train, y_train)
```

```python
# Exportar el árbol de decisión a Graphviz
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None,
                                feature_names=X.columns,
                                class_names=y.unique().astype(str),
                                filled=True, rounded=True,
                                special_characters=True)
graph = graphviz.Source(dot_data)

graph.render("decision_tree")  # se guarda el gráfico como un archivo PDF
graph # Esto abrirá el gráfico en el visor de Graphviz
```

A menor índice de Gini mayor pureza de la clase que se esta agrupando.
Se realizo un árbol con una profundidad igual a 4 pero evaluando las metricas, el rendimiento del algoritmo era menor que cuando la profundidad es igual a 6.

```python
# Se generan predicciones con los datos de prueba (X_test)
y_pred = clf.predict(X_test)

print(y_pred)
```

###Métricas

Hay 854 respuestas del modelo que dicen que los pacientes no tienen riesgo a padecer enfermedades coronarias y realmente es así según la base de datos. Sin embargo, 134 respuestas del modelo predicen que los pacientes no padecen el riesgo aunque en verdad si lo tienen.

Por otro lado, 15 respuestas del modelo predicen el riesgo a padecer enfermedades coronarias lo que se condice con la realidad, pero hay 14 respuestas que predicen que los pacientes padecen el riesgo lo que no concuerda con la realidad.
